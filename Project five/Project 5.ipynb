```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Import data
a = pd.read_csv('F:/Work/Visualizatoin Code with Python/Free code camp data analysis with python projects/Project five/epa-sea-level.csv')
df = pd.DataFrame(a)
```

```python
fig, ax = plt.subplots(figsize = (9,6))
plt.scatter( x = df['Year'], y = df['CSIRO Adjusted Sea Level'] )

result = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
v = list(df.Year)
v = v + [i for i in range(2014, 2051)]
x = pd.Series(v)
y = result.intercept + (result.slope * x ) 

ax.plot(x , y, color ='red' )

'''
point_one = result.intercept + (result.slope * 1880 )
point_two = result.intercept + (result.slope * 2050 )
'''

ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title("Rise in Sea Level")

result_two = linregress(x = df.loc[df.Year >= 2000 , 'Year' ] , y = df.loc[df.Year >= 2000 , 'CSIRO Adjusted Sea Level' ])
x2 = x[x >= 2000]
y2 = result_two.intercept + (result_two.slope * x2 )

'''
point_Rone = result_two.intercept + (result_two.slope * 2000 )
point_Rtwo = result_two.intercept + (result_two.slope * 2050 )
'''


ax.plot(x2, y2 , color = 'green')
```

```python
# Creating Predictor as a Function

def draw_plot():
    a = pd.read_csv('F:/Work/Visualizatoin Code with Python/Free code camp data analysis with python projects/Project five/epa-sea-level.csv')
    df = pd.DataFrame(a)
    

    # Create scatter plot
    fig, ax = plt.subplots(figsize = (9,6))
    plt.scatter(x = df['Year'], y= df['CSIRO Adjusted Sea Level'] )


    # Create first line of best fit
    result = linregress(x = df['Year'], y= df['CSIRO Adjusted Sea Level'])
    v = list(df.Year)
    v = v + [i for i in range(2014, 2051)]
    x = pd.Series(v)
    y = result.intercept + (result.slope * x ) 
    
    ax.plot(x , y, color ='red' )


    # Create second line of best fit
    result_two = linregress(x = df.loc[df.Year >= 2000 , 'Year' ] , y = df.loc[df.Year >= 2000 , 'CSIRO Adjusted Sea Level' ])
    x2 = x[x >= 2000]
    y2 = result_two.intercept + (result_two.slope * x2 )

    
    ax.plot(x2, y2 , color = 'green')


    # Add labels and title
    
    ax.set_xlabel('Year')
    ax.set_ylabel('Sea Level (inches)')
    ax.set_title("Rise in Sea Level")

    
    # Save plot and return data for testing (DO NOT MODIFY)
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```

```python
draw_plot()
```
